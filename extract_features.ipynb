```python
from skimage.io import imread
from sklearn.externals import joblib
import glob
import os
import imutils
import warnings
from hog import get_hog
import cv2
from configuration import *
# list = os.listdir(total_pos_im_path) # check pos directory path
# number_files = len(list)
# print(number_files)
# list = os.listdir(total_neg_im_path) # check neg directory path
# number_files = len(list)
# print(number_files)
def extract_features():
    # If feature directories don't exist, create them
    if not os.path.isdir(train_pos_feat_ph):
        os.makedirs(train_pos_feat_ph)
    # If feature directories don't exist, create them
    if not os.path.isdir(train_neg_feat_ph):
        os.makedirs(train_neg_feat_ph)

    for im_path in glob.glob(os.path.join(total_pos_im_path,"*")): # load positive and process features
        print("+", end ="")
        im = cv2.imread(im_path, 0)
        im = imutils.resize(im,width=64)
        try:
            feature = get_hog(im,csize=8,bin=9)#hog(im, orientations, pixels_per_cell, cells_per_block, visualize=visualize)
        except:
            warnings.warn("Error in extracting Positive Feature")
        feat_name = os.path.split(im_path)[1].split(".")[0] + ".feat"
        feat_path = os.path.join(train_pos_feat_ph, feat_name)
        joblib.dump(feature, feat_path)
    print("Pos features saved in",train_pos_feat_ph)
    for im_path in glob.glob(os.path.join(total_neg_im_path,"*")): # load negative and process features
        print("-", end ="")
        im = cv2.imread(im_path, 0)
        im = imutils.resize(im,width=64)
        try:
            feature=get_hog(im,csize=8,bin=9)#hog(im, orientations, pixels_per_cell, cells_per_block, visualize=visualize)
        except:
            warnings.warn("Error in extracting Negative Feature")
        fd_name = os.path.split(im_path)[1].split(".")[0] + ".feat"
        fd_path = os.path.join(train_neg_feat_ph, fd_name)
        joblib.dump(feature, fd_path)
    print("Neg features saved in",total_neg_im_path)
    print("Completed calculating features from images")
extract_features()
```
